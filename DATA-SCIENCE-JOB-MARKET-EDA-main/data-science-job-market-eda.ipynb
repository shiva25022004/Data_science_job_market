## 6.Converting Salary range to minimum and maximum salary


```python
def extract_salary(salary_range):
    a,b=salary_range.split("-")
    a=int(a.replace('$',"").replace('K',""))
    b=int(b.replace('$',"").replace('K',""))
    return a,b
```

```python
salary_components = df3["Salary Range"].apply(extract_salary)
df3[["Min Salary", "Max Salary"]] = pd.DataFrame(salary_components.tolist(), index=df3.index)
```

## 6.Converting experience to minimum and maximum experience



```python
def extract_experience(exp):
    a=exp.split(" ")
    #print(a)
    return int(a[0]),int(a[2])
```

```python
exp_components = df3['Experience'].apply(extract_experience)
df3[["Min Experience", "Max Experience"]] = pd.DataFrame(exp_components.tolist(), index=df3.index)
```

## 13.Extracting Skills using NLP

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

def extract_skills(job_description):
    # Process the job description text
    doc = nlp(job_description)

    # List to store extracted skills
    skills = []

    # Iterate through each token in the document
    for token in doc:
        # Check if the token is a noun and has a dependency relation of compound
        if token.pos_ == "NOUN" and token.dep_ == "compound":
            # Append the compound noun and its head (usually a skill)
            skills.append(token.text + " " + token.head.text)

        # Check if the token is a noun and has a dependency relation of apposition
        elif token.pos_ == "NOUN" and token.dep_ == "appos":
            # Append the appositive and its head (usually a skill)
            skills.append(token.text + " " + token.head.text)

        # Check if the token is a noun and its head is a prepositional object
        elif token.pos_ == "NOUN" and token.head.dep_ == "pobj":
            # Append the noun and its preposition (usually a skill)
            skills.append(token.text + " " + token.head.head.text)

    z=""
    for i in skills[1:]:
        z=z+i+','
    return z
```

```python
def merge(row):
    return row['Job Description'] + " " + row['Responsibilities'] + " " + row['skills']

# Apply the merge function to each row of the DataFrame
df4['combined'] = df4.apply(merge, axis=1)


df4['Skills Found']=df4['combined'].apply(extract_skills)
```

## 14. Skills founnd vs frequency plot

```python
import matplotlib.pyplot as plt
all_skills = df4['Skills Found'].str.split(',').sum()

# Count the occurrences of each skill
skill_counts = pd.Series(all_skills).value_counts()

# Plot the results
plt.figure(figsize=(10, 6))
skill_counts.plot(kind='bar')
plt.title('Skills Found in Job Descriptions')
plt.xlabel('Skills')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
